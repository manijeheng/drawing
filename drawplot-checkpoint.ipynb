```python
import matplotlib.pyplot  as plt
from random import randint as rnd
```

```python
x = [2,4,7,2]
y = [3,7,4,3]
plt.plot(x,y)
plt.show
```

```python
x = list(range(50))
y = [rnd(-10,10) for i in range(50)]
z = [rnd(-10,10) for i in range(50)]
plt.figure(figsize = (10,5))
plt.title("random")
plt.xlabel("ea")
plt.ylabel("time")

plt.grid()
plt.plot(x,y,label = "gasoline",color = "red")
plt.plot(x,z,label = "oil")
plt.legend()
```
